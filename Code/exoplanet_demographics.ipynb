```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rc('xtick', labelsize=27) 
matplotlib.rc('ytick', labelsize=27)
```

```python
data = pd.read_csv('exoplanet_data_sun-like.csv')
data_all = pd.read_csv('exoplanet_data.csv')
```

```python
print(data.columns)
```

```python
periods = data['pl_orbper']
masses = data['pl_masse']


periods_all = data_all['pl_orbper']
masses_all = data_all['pl_masse']
```

```python
m_earth = 1.
p_earth = 365.25
```

```python
m_jup = 317.83
p_jup = 4327.631
```

```python
from matplotlib.offsetbox import OffsetImage, AnnotationBbox

def getImage(path):
    return OffsetImage(plt.imread(path))
```

```python
fig, ax = plt.subplots(1, 1, figsize = [15,15])
ax.plot(periods, masses, 'o', color = "#7b3e65", alpha = 0.3, markersize = 13)


ax.set_yscale('log')
ax.set_xscale('log')




paths = [
    'earth.png',
    'jupiter.png',
    ]

y = [m_earth,m_jup]
x = [p_earth,p_jup]

ax.scatter(x, y) 

for x0, y0, path in zip(x, y, paths):
    ab = AnnotationBbox(getImage(path), (x0, y0), 
                        frameon=False)
    ax.add_artist(ab)
    

ax.set_xlim(.1, 100000)
ax.set_ylim(.1, 10000)

ax.set_ylabel(r'mass [M$_{Earth}$]', fontsize = 36)
ax.set_xlabel('orbital period [days]', fontsize = 36)

ax.set_title('confirmed exoplanets around Sun-like stars', fontsize = 36)
fig.savefig('final_figures/exoplanets_sun-like.pdf')

fig.tight_layout()
```
