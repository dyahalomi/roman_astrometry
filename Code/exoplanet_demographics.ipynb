```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('exoplanet_data.csv')
```

```python
print(data.columns)
```

```python
periods = data['pl_orbper']
masses = data['pl_masse']
```

```python
m_earth = 1.
p_earth = 365.25
```

```python
m_jup = 317.83
p_jup = 4327.631
```

```python
from matplotlib.offsetbox import OffsetImage, AnnotationBbox

def getImage(path):
    return OffsetImage(plt.imread(path))
```

```python
fig, ax = plt.subplots(1, 1, figsize = [15,15])
ax.plot(periods, masses, 'o')
#ax.plot(p_earth, m_earth, 'o', color = 'green', markersize = 100)
#ax.plot(p_jup, m_jup, 'o', color = 'orange', markersize = 10)

ax.set_yscale('log')
ax.set_xscale('log')




paths = [
    'earth.png',
    'jupiter.png',
    ]

y = [m_earth,m_jup]
x = [p_earth,p_jup]

ax.scatter(x, y) 

for x0, y0, path in zip(x, y, paths):
    ab = AnnotationBbox(getImage(path), (x0, y0), frameon=False, fontsize=0.01)
    ax.add_artist(ab)

```
